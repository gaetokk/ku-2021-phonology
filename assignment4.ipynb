```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
def phasor_sine(amp, sr, dur, freq, phase):
  t = np.arange(1, sr * dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  s = amp * np.sin(theta)
  return s, t, theta
```

```python
def getFrequency(note):
    A4=440
    notes = ['A', 'A#', 'B', 'C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#']

    octave = int(note[2]) if len(note) == 3 else int(note[1])
        
    keyNumber = notes.index(note[0:-1]);
    
    if (keyNumber < 3) :
        keyNumber = keyNumber + 12 + ((octave - 1) * 12) + 1; 
    else:
        keyNumber = keyNumber + ((octave - 1) * 12) + 1; 

    return A4 * 2** ((keyNumber- 49) / 12)
```

```python
amp = 1; sr = 10000; phase = 0

A5 = getFrequency('A5')
Ab5 = getFrequency('G#5')
B5 = getFrequency('B5')
Db6 = getFrequency('C#6')
E5 = getFrequency('E5')
Gb5 = getFrequency('F#5')
```

```python
notes = [E5, Db6, 0, B5, 0, A5, 0, Ab5, Ab5, Gb5, E5, E5, 0, \
          E5, E5, E5, Db6, 0, B5, 0, A5, A5, Gb5, E5, Gb5, E5, E5, 0]
durs = [0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.25, 0.5, 0.25, 0.5, 1.75, \
        0.25, 0.25, 0.75, 0.5, 0.5, 0.5, 0.5, 0.75, 0.25, 0.5, 0.25, 0.75, 0.25, 0.5, 1.75]
```

```python
s_all = np.empty(0)

for note, dur in zip(notes, durs):
  s, t, th= phasor_sine(amp, sr, dur, note, phase)
  s_all = np.concatenate((s_all, s))
```

```python
ipd.Audio(s_all, rate=sr)
```

```python
import scipy.io.wavfile
scipy.io.wavfile.write("dynamite.wav", sr, s_all)
```
